Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans. Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of customizable plots such as line plots, scatter plots, bar plots, histograms, heatmaps, and many more. Matplotlib was created by John D. Hunter in 2003 and is now one of the most widely used data visualization libraries in Python.

Matplotlib provides a flexible and powerful framework for creating high-quality visualizations. It can be used for a variety of purposes such as data exploration, communication of insights, teaching and learning, and automating repetitive tasks. Matplotlib is also highly extensible, with a large number of plugins and third-party packages available to enhance its functionality.

Matplotlib is built on top of the NumPy and SciPy libraries, and is compatible with many other Python libraries such as Pandas and Seaborn. It can be used in a variety of environments such as Jupyter notebooks, web applications, and standalone Python scripts. Overall, Matplotlib is a versatile and powerful library for creating high-quality visualizations in Python.

The Pyplot module of Matplotlib provides a convenient interface for creating a variety of plots. Here are five types of plots that can be plotted using Pyplot:

1. Line Plot - Used to visualize the relationship between two variables.

2. Scatter Plot - Used to visualize the relationship between two variables as a set of points.

3. Bar Plot - Used to visualize the distribution of a categorical variable.

4. Histogram - Used to visualize the distribution of a numerical variable.

5. Box Plot - Used to visualize the distribution of a numerical variable and identify outliers.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.



```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

Bar plots are commonly used to compare the distribution of a categorical variable across different groups or to visualize the frequency of a categorical variable.

Bar plots can be created using the bar() function in the Matplotlib library. 

Bar plots are useful for several reasons:

1. Comparing groups - Bar plots are an effective way to compare the frequency or proportion of a categorical variable across different groups.

2. Showing distributions - Bar plots can be used to visualize the distribution of a categorical variable, especially when the number of categories is small.

3. Highlighting differences - Bar plots can be used to highlight differences between categories or groups, making it easier to identify trends or patterns.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans. A box plot is a type of plot used to visualize the distribution of a dataset. It displays the five-number summary of a dataset (minimum, first quartile, median, third quartile, and maximum) as well as any outliers. The box represents the interquartile range (IQR) of the data, which is the distance between the first and third quartiles. The whiskers extend from the box to the minimum and maximum values that are not considered outliers, while outliers are plotted individually as points outside the whiskers.

Box plots can be created using the boxplot() function in the Matplotlib library.

Box plots are useful for several reasons:

1. Visualizing distributions - Box plots provide a quick visual summary of the distribution of a dataset, including the central tendency, variability, and skewness.

2. Comparing distributions - Box plots are an effective way to compare the distribution of a dataset across different groups or conditions.

3. Identifying outliers - Box plots can help to identify outliers, which are values that are unusually large or small compared to the rest of the data.

4. Communicating results - Box plots are a common and widely recognized visualization technique, making them a useful tool for communicating results to others.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# create the box plot
plt.boxplot(data)

# add labels and title
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box Plot Example')

# show the plot
plt.show()
```
